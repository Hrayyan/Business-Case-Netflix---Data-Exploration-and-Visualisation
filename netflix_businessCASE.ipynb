COMPARISON BETWEEN TV SHOWS AND MOVIES

```python
data=country_explode.merge(netflix_df[['show_id','type']],how='left', left_on='show_id',right_on='show_id')
data
```

```python
data[data['type']=='Movie'].groupby('country').count().sort_values(by='show_id',ascending=False)
```

```python
data[data['type']=='TV Show'].groupby('country').count().sort_values(by='show_id',ascending=False)
```

```python
movies_df = data[data['type'] == 'Movie']

# Step 2: Group by 'country' and count movies
country_movie_counts = movies_df['country'].value_counts()

# Step 3: Convert to DataFrame for better manipulation
country_movie_counts_df = pd.DataFrame({'Country': country_movie_counts.index, 'Number of Movies': country_movie_counts.values})

# Step 4: Sort and get the top 10 countries
top_10_countries = country_movie_counts_df.head(10)

# Print the top 10 countries and their movie counts
print(top_10_countries)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries['Country'], top_10_countries['Number of Movies'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Number of Movies')
plt.title('Top 10 Countries by Number of Movies Produced on Netflix')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
```

```python
tv_shows_df = data[data['type'] == 'TV Show']

# Step 2: Group by 'country' and count TV shows
country_tv_show_counts = tv_shows_df['country'].value_counts()

# Step 3: Convert to DataFrame for better manipulation
country_tv_show_counts_df = pd.DataFrame({'Country': country_tv_show_counts.index, 'Number of TV Shows': country_tv_show_counts.values})

# Step 4: Sort and get the top 10 countries
top_10_countries_tv = country_tv_show_counts_df.head(10)

# Print the top 10 countries and their TV show counts
print(top_10_countries_tv)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries_tv['Country'], top_10_countries_tv['Number of TV Shows'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Number of TV Shows')
plt.title('Top 10 Countries by Number of TV Shows Produced on Netflix')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
```

BEST TIME TO LAUNCH A TV SHOW AND MOVIES

```python
tv_shows = netflix_df[netflix_df['type'] == 'TV Show']
tv_shows['date_added'] = pd.to_datetime(tv_shows['date_added'])  # Convert 'date_added' to datetime

# Group by week and count TV shows added each week
tv_show_weekly_counts = tv_shows.groupby(tv_shows['date_added'].dt.week)['show_id'].count()

tv_show_weekly_counts.head()
```

```python
best_week_for_tv_shows = tv_show_weekly_counts.idxmax()
print("Best Week to Release TV Shows (Week Number):", best_week_for_tv_shows)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(tv_show_weekly_counts.index, tv_show_weekly_counts.values, marker='o', linestyle='-', color='green', label='TV Shows')
plt.xlabel('Week Number')
plt.ylabel('Number of TV Shows')
plt.title('Number of TV Shows Released Each Week')
plt.legend()
plt.grid(True)
plt.tight_layout()
```

```python
movies_daf = netflix_df[netflix_df['type'] == 'Movie']
movies_daf['date_added'] = pd.to_datetime(movies_daf['date_added'])  # Convert 'date_added' to datetime

# Group by week and count movies added each week
movie_weekly_counts = movies_daf.groupby(movies_daf['date_added'].dt.week)['show_id'].count()
movie_weekly_counts.head()
```

```python
best_week_for_movies = movie_weekly_counts.idxmax()
print("Best Week to Release Movies (Week Number):", best_week_for_movies)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(movie_weekly_counts.index, movie_weekly_counts.values, marker='o', linestyle='-', color='green', label='TV Shows')
plt.xlabel('Week Number')
plt.ylabel('Number of Movies')
plt.title('Number of Movies Released Each Week')
plt.legend()
plt.grid(True)
plt.tight_layout()
```

Analysis of actors/directors

```python
data1= director_explode.merge(netflix_df[['show_id','type']], how='left', left_on='show_id',right_on='show_id')
data1
```

```python
# Step 1: Filter for rows where the 'director' column is not missing
data1 = data1.dropna(subset=['director'])

# Step 2: Filter out rows where the 'director' is 'Unknown' or 'NA' (customize as needed)
data1 = data1[~data1['director'].str.lower().isin(['Unknown director', 'na'])]

# Step 3: Separate movies and TV shows
movies_df = data1[data1['type'] == 'Movie']
tv_shows_df = data1[data1['type'] == 'TV Show']

# Step 4: Group by 'director' and count the number of movies for each director
movie_director_counts = movies_df.groupby('director')['show_id'].count()

# Step 5: Group by 'director' and count the number of TV shows for each director
tv_show_director_counts = tv_shows_df.groupby('director')['show_id'].count()

# Step 6: Sort the director counts in descending order and get the top 10 directors for movies and TV shows
top_10_directors_movies = movie_director_counts.nlargest(10)
top_10_directors_tv_shows = tv_show_director_counts.nlargest(10)

# Print the top 10 directors for movies and TV shows
print("Top 10 Directors for Movies:")
print(top_10_directors_movies)

print("\nTop 10 Directors for TV Shows:")
print(top_10_directors_tv_shows)
```

```python
# Create a bar plot for top 10 directors in movies
plt.figure(figsize=(12, 6))
plt.bar(top_10_directors_movies.index, top_10_directors_movies.values, color='skyblue')
plt.xlabel('Director')
plt.ylabel('Number of Movies')
plt.title('Top 10 Directors in Movies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Create a bar plot for top 10 directors in TV shows
plt.figure(figsize=(12, 6))
plt.bar(top_10_directors_tv_shows.index, top_10_directors_tv_shows.values, color='lightcoral')
plt.xlabel('Director')
plt.ylabel('Number of TV Shows')
plt.title('Top 10 Directors in TV Shows')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
